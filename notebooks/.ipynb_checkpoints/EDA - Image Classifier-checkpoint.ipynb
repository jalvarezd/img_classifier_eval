````python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
````

# Exploratory Data Analysys
Antes de realizar cualquier proceso de entrenamiento, es necesario realizar un EDA, con el fin de determinar como esta distribuido nuestro dataset, para esto se utiliza la funcion ```get_distribution```que nos permite visualizar las diferentes categorias de nuestro dataset y la cantidad de items presentes en cada una de ellas


````python
# Get classes from raw dataset

def get_dir_list(src_path = "."):
        """
        Description: Obtain a list of the subdirectories from a given path

        Arguments:  - src_path [Default="."]: path from where we want to obtain a list of subdirectories

        Return:     - _folder_list: list of subdirectories names
        """
        _folder_list = [f.name for f in os.scandir(src_path) if f.is_dir()]

        return _folder_list

def get_files_from_dir_ext(src_path=".", ext=".jpeg"):
        """
        Description: Obtain a list of the any kind of files from a given directory.

        Arguments:   - src_path [Default ="."]: path from where we will scan audiofiles.
                     - ext [Default=".txt"] = filename extension
        Return:      - _files: list of files with the given extension
        """
        _src_path = src_path
        _ext = ext
        _files = []
        with os.scandir(_src_path) as entries:
            for entry in entries:
                if (entry.name).endswith(_ext):
                    _files.append(entry.name)

        return _files
    
def get_distribution(src_path, labels ,plot=False):
        """
        # Description: Obtain a list of the class ditribution from a given dataset

        # Arguments: src_path: path to the the dataset that will be analiyzed
                     labels: list of the labels to analyze
                     plot (defaul = False): if True, we will plot an axis bar graph with the amount of
                                            items per class in our dataset
                                            if False, this function will return only a list contaning the
                                            item distribution per class
        """
        _src_path = src_path

        
        _distribution = []
        _total_files = 0
        _numbers = []
        _labels = labels

        for _label in _labels:
            _files = get_files_from_dir_ext(os.path.join(_src_path, _label), ".jpeg")
            _sub_dis = [_label, len(_files)]
            _total_files = _total_files + len(_files)
            _numbers.append(len(_files))
            _distribution.append(_sub_dis)

        if plot == False:

            return _total_files, _distribution, _numbers
        elif plot == True:
            fig = plt.figure(figsize=(20,5))
            plt.title("Dataset Ditribution")
            ax = fig.add_axes([0,0,1,1])
            plt.xlabel("class")
            plt.ylabel("instances")
            ax.bar(_labels, _numbers)
            plt.xticks(rotation=45)
            #plt.show()

            return _total_files, _distribution, _numbers, fig
        
# Obtener la informacion del dataset RAW
RAW_DATA_FOLDER = "../data/raw"
labels = get_dir_list(RAW_DATA_FOLDER)
#print(labels)

total_files, distribution, numbers, fig = get_distribution(RAW_DATA_FOLDER, labels, True)

print("dataset info:\ntotal files: {}\ndistribution: {}".format(total_files, distribution))
````

como se puede apreciar en la imagen superior, el dataset no se encuentra balanceado, esto puede ser un problema al momento de entrenar un modelo clasificador de imagenes. Para atacar este problema se pueden seguir las siguientes estrategias:

1. realizar data augmentation en todo el dataset de entrenamiento
2. generar imagenes sinteticas en aquellas categorias donde se tienen pocos items 

En ambos casos hay que tener en cuenta que el tamao del dataset es bastante pequeo para aplicaciones de Deep Learning por lo que puede tener problemas de overfitting independientemente de que empleen las dos estrategias propuestas


# Similitudes entre clases:
Otro aspecto a tener en cuenta es que tan similares son las imagenes de diferentes clases entre si, en este caso al tener pocas categorias se puede visualizar algunas imagenes del dataset, en casos donde se tengan muchas categorias es conveniente utilizar un algoritmo de clusttering para saber que tan correlacionado esta nuestro dataset.

````python
# Plot images
fig = plt.figure(tight_layout=True, figsize=(30,30))
#plot images:
plot_count = 1
for label in labels[0:12]:
    # get images
    label_path = os.path.join(RAW_DATA_FOLDER, label)
    img_files = get_files_from_dir_ext(label_path, ".jpeg")
    
    for img in img_files[0:3]:
        #print(plot_count)
        ax = fig.add_subplot(len(labels),3, plot_count)
        image = plt.imread(os.path.join(label_path,img))
        ax.set_title(label)
        ax.imshow(image)
        plot_count = plot_count + 1

````

observando las imagenes de las diferentes categorias, es claro que se aunque comparten caracteristicas sobretodo en el color, la forma y contorno de cada uno es bastante diferenciable entre clases, lo cual facilitaria en parte la clasificacion de imagenes
