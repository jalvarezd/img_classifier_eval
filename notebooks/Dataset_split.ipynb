# Generacion del dataset de entrenamiento 
En el presente notebook se indica como se realizan dos tipos de "balanceo" del dataset de entrenamiento:
1. Tomando el valor minimo de items para posteriormente hacer data augmentation durante el entrenamiento
2. Creando imagenes sinteticas en los datasets con pocas muestras

para esto utilizaremos las mismas funciones del EDA 

````python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Get classes from raw dataset

def get_dir_list(src_path = "."):
        """
        Description: Obtain a list of the subdirectories from a given path

        Arguments:  - src_path [Default="."]: path from where we want to obtain a list of subdirectories

        Return:     - _folder_list: list of subdirectories names
        """
        _folder_list = [f.name for f in os.scandir(src_path) if f.is_dir()]

        return _folder_list

def get_files_from_dir_ext(src_path=".", ext=".jpeg"):
        """
        Description: Obtain a list of the any kind of files from a given directory.

        Arguments:   - src_path [Default ="."]: path from where we will scan audiofiles.
                     - ext [Default=".txt"] = filename extension
        Return:      - _files: list of files with the given extension
        """
        _src_path = src_path
        _ext = ext
        _files = []
        with os.scandir(_src_path) as entries:
            for entry in entries:
                if (entry.name).endswith(_ext):
                    _files.append(entry.name)

        return _files
    
def get_distribution(src_path, labels ,plot=False):
        """
        # Description: Obtain a list of the class ditribution from a given dataset

        # Arguments: src_path: path to the the dataset that will be analiyzed
                     labels: list of the labels to analyze
                     plot (defaul = False): if True, we will plot an axis bar graph with the amount of
                                            items per class in our dataset
                                            if False, this function will return only a list contaning the
                                            item distribution per class
        """
        _src_path = src_path

        
        _distribution = []
        _total_files = 0
        _numbers = []
        _labels = labels

        for _label in _labels:
            _files = get_files_from_dir_ext(os.path.join(_src_path, _label), ".jpeg")
            _sub_dis = [_label, len(_files)]
            _total_files = _total_files + len(_files)
            _numbers.append(len(_files))
            _distribution.append(_sub_dis)
            _fig = None

        if plot == False:

            return _total_files, _distribution, _numbers, _fig
        elif plot == True:
            fig = plt.figure(figsize=(20,5))
            plt.title("Dataset Ditribution")
            ax = fig.add_axes([0,0,1,1])
            plt.xlabel("class")
            plt.ylabel("instances")
            ax.bar(_labels, _numbers)
            plt.xticks(rotation=45)
            #plt.show()

            return _total_files, _distribution, _numbers, fig
        
````

````python
# GET DATASET INFO 
RAW_DATA_FOLDER = "../data/raw"

labels = get_dir_list(RAW_DATA_FOLDER)

total_files, distribution, numbers, fig = get_distribution(RAW_DATA_FOLDER, labels, True)

print("dataset info:\ntotal files: {}\ndistribution: {}".format(total_files, distribution))
````

# Creacion de directorios para almacenar nuestro dataset de entrenamiento
La siguiente funcion sirve para generar un dataset de entrenamiento con la siguiente estructura: 
```txt
training_dataset/
    training
        ...class_a/
        ......a_image_1.jpg
        ......a_image_2.jpg
        ...class_b/
        ......b_image_1.jpg
        ......b_image_2.jpg
    validation
        ...class_a/
        ......a_image_11.jpg
        ......a_image_22.jpg
        ...class_b/
        ......b_image_11.jpg
        ......b_image_22.jpg
    testing
        ...class_a/
        ......a_image_31.jpg
        ......a_image_32.jpg
        ...class_b/
        ......b_image_31.jpg
        ......b_image_32.jpg
```

````python
# create a trainig dataset:
def generate_training_directory(train_dst_path, labels):
    _train_dst_path = train_dst_path
    _labels = labels
    if os.path.isdir(_train_dst_path):
        print("directory already exists")
        _status = False
        return _status
    else:
        # Create directory without asking to remove
        _subdirs = ["training", "validation", "testing"]
        try:
            # Create main dataset directory:
            os.mkdir(_train_dst_path)
            for _subdir in _subdirs:
                _subdir_path = os.path.join(_train_dst_path, _subdir)
                os.mkdir(_subdir_path)
                for _label in _labels:
                    os.mkdir(os.path.join(_subdir_path, _label))
                    print("directory {} ready".format(os.path.join(_subdir_path, _label)))
            _status = True
            return  _status
        except OSError:
            print("Directory {} failed to create".format(_train_dst_path))
            _status = False
            return _status
````

````python
train_dataset_name = ["training_dataset_bal_min", "training_dataset_bal_extended"]

for name in train_dataset_name:
    train_dst_path = os.path.join("../data/train", name)
    generate_training_directory(train_dst_path, labels)
````

# Generando un dataset con valores minimos
La estrategia empleada aqui es balancear el dataset usando como filtro el valor mas bajo de muestras presentes en una clase, y posteriormente utilizar data augmentation durante el entrenamiento para evitar el overfitting

````python
# Dataset Distribution data
# GET DATASET INFO 
RAW_DATA_FOLDER = "../data/raw"

labels = get_dir_list(RAW_DATA_FOLDER)
total_files, distribution, numbers, fig = get_distribution(RAW_DATA_FOLDER, labels, True)

distribution
````

````python
# Split dataset:
import shutil

# Use the class with the minimum amount of files to split the remain classes 

min_files_training = int(171*0.8) #80% of the minimum value to training
min_files_validation = int(171*0.1)#10% of the minimum value to validation

print(min_files_training)
print(min_files_validation)

train_dst_path = os.path.join("../data/train", train_dataset_name[0])

for dist in distribution:
    src_dataset_path = "../data/raw"
    #get label:
    label = dist[0]
    total_files = dist[1]
    # get files:
    files = get_files_from_dir_ext(os.path.join(src_dataset_path, label),".jpeg")
    
    #copy to training dataset:
    for file in files[0:min_files_training]:
        dst_path = os.path.join(os.path.join(train_dst_path, "training"),label)
        file_path = os.path.join(os.path.join(src_dataset_path, label), file)
        print(file_path)
        shutil.copyfile(file_path, os.path.join(dst_path, file))
    
    #copy to validation dataset:
    for file in files[min_files_training:(min_files_training+min_files_validation)]:
        dst_path = os.path.join(os.path.join(train_dst_path, "validation"),label)
        file_path = os.path.join(os.path.join(src_dataset_path, label), file)
        shutil.copyfile(file_path, os.path.join(dst_path, file))
    
    #copy to testing dataset:
    for file in files[(min_files_training+min_files_validation):]:
        dst_path = os.path.join(os.path.join(train_dst_path, "testing"),label)
        file_path = os.path.join(os.path.join(src_dataset_path, label), file)
        shutil.copyfile(file_path, os.path.join(dst_path, file))
````

````python
# plot training dataset split:
train_split_path = os.path.join(train_dst_path, "training")
train_total_files, train_distribution, train_numbers, train_fig = get_distribution(train_split_path, labels, True)

# plot validation dataset split:
train_split_path = os.path.join(train_dst_path, "validation")
train_total_files, train_distribution, train_numbers, train_fig = get_distribution(train_split_path, labels, True)

# plot testing dataset split:
train_split_path = os.path.join(train_dst_path, "testing")
train_total_files, train_distribution, train_numbers, train_fig = get_distribution(train_split_path, labels, True)
````
